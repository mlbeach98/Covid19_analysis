As there are dates associated with the data and I have plenty of experience working with stock market data, the first thing I will do is look at this from a time series perspective.

```python
nyCovid.dtypes
```

In our case our dataset is already sorted by date, but just to cover our bases lets resort by date. To do this we will first have to change the datatype of the "ObservationDate" column to datetime from a general object.

```python
nyCovid['ObservationDate'] = pd.to_datetime(nyCovid['ObservationDate'], format='%m/%d/%Y')
```

```python
nyCovid.dtypes
```

Since our dataset is already sorted by date, we can first test this method of sorting by sorting descending first and looking at the top of our dataframe. After properly sorting we will also reset our index, but we will keep the "SNo" column to reference the true row location in the main dataset if we ever need to come back to reference it.

```python
nyCovid.sort_values(by=["ObservationDate"], ascending=False, inplace=True)
```

```python
nyCovid.head()
```

```python
nyCovid.sort_values(by=["ObservationDate"], inplace=True)
```

```python
nyCovid.reset_index(inplace=True, drop=True)
```

```python
nyCovid.head()
```

Now that we have our sorted dataset, we will plot the confirmed cases in New York.

```python
plt.figure(figsize = (15, 8))
ax = sns.lineplot(x = "ObservationDate", y = "Confirmed", data = nyCovid)
ax.set_ylabel("Confirm Covid Cases", fontsize = 15)
ax.set_xlabel("Date", fontsize = 15)
ax.set_title("New York State Confirmed Covid Cases", fontsize = 20)
```

This plot gives us a good initial view of the rise in Covid cases from March 2020 through March 2021

The next thing I'm interested to see is the trend for recovered cases and deaths alongside the confirmed cases. For this to be able to plot easily, I will flatten the dataset and keep only the relevant data for this plot.

```python
flatNy = pd.DataFrame(columns = ['ObservationDate', 'Type', 'Count'])
for index, row in nyCovid.iterrows():
    confDict = {}
    confDict['ObservationDate'] = row['ObservationDate']
    confDict['Type'] = "Confirmed"
    confDict['Count'] = row['Confirmed']
    
    recDict = {}
    recDict['ObservationDate'] = row['ObservationDate']
    recDict['Type'] = "Recovered"
    recDict['Count'] = row['Recovered']
    
    deathDict = {}
    deathDict['ObservationDate'] = row['ObservationDate']
    deathDict['Type'] = "Deaths"
    deathDict['Count'] = row['Deaths']
    
    flatNy = flatNy.append(confDict, ignore_index = True)
    flatNy = flatNy.append(recDict, ignore_index = True)
    flatNy = flatNy.append(deathDict, ignore_index = True)
```

```python
flatNy.head()
```

Now that we have a flattened dataset, we can plot all three counts alongside one another.

```python
plt.figure(figsize = (15, 8))
ax = sns.lineplot(x = "ObservationDate", y = "Count", hue="Type", data = flatNy)
ax.set_ylabel("Count", fontsize = 15)
ax.set_xlabel("Date", fontsize = 15)
ax.set_title("New York State Covid Cases (Confirmed/Recovered/Deaths)", fontsize = 20)
```

Plotting this data as is, we don't get a lot of valuable information out of the deaths and recovered data, with the recovered data looking almost perfectly flat. Taking a look at a graph of the recovered data by itself we can see that it is flat, with no recovered cases reported in New York (although this clearly isn't the case).

```python
plt.figure(figsize = (15, 8))
ax = sns.lineplot(x = "ObservationDate", y = "Recovered", data = nyCovid)
ax.set_ylabel("Recovered Covid Cases", fontsize = 15)
ax.set_xlabel("Date", fontsize = 15)
ax.set_title("New York State Recovered Covid Cases", fontsize = 20)
```
